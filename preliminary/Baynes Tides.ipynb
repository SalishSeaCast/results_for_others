Tides in Baynes Sound, Oct 18, 2002 for Debby Ianson

```python
%matplotlib inline
from matplotlib import pylab
import matplotlib.pyplot as plt
import netCDF4 as NC
import numpy as np
```

```python
fT = NC.Dataset('../Oct18OneDay/SalishSea_1h_20021018_20021018_grid_T.nc','r')
temp = fT.variables['votemper'][:]
```

```python
fU = NC.Dataset('../Oct18OneDay/SalishSea_1h_20021018_20021018_grid_U.nc','r')
uvel = fU.variables['vozocrtx'][:]
mu = uvel == 0
uvel_masked = np.ma.array(uvel,mask=mu)
print uvel.shape
```

```python
fV = NC.Dataset('../Oct18OneDay/SalishSea_1h_20021018_20021018_grid_V.nc','r')
vvel = fV.variables['vomecrty'][:]
mv = vvel == 0
vvel_masked = np.ma.array(vvel,mask=mv)
```

```python
level = 0
jmin = 115
jmax = 145
imin = 585
imax = 645
lmin=3; lmax=10;
print "One grid point is 0.25 m/s"
for time in np.arange(0,24,2):
    plt.figure(figsize=(9,9))
    plt.subplot(1,2,1)
    plt.quiver(0.5*uvel_masked[time,level,imin:imax-1,jmin:jmax-1]+
           0.5*uvel_masked[time,level,imin+1:imax+1-1,jmin:jmax-1],
           0.5*vvel_masked[time,level,imin:imax-1,jmin:jmax-1]+
           0.5*vvel_masked[time,level,imin:imax-1,jmin+1:jmax+1-1],color='b',scale=4,pivot='middle')
    tempPlot = temp[time,level,:,:]
    land=tempPlot
    mu =land != 0
    land= np.ma.array(land,mask=mu)
    pylab.pcolor(np.arange(1,jmax-jmin)-1,np.arange(1,imax-imin)-1,
             land[imin+1:imax,jmin+1:jmax],vmin=lmin,vmax=lmax,cmap='hot')
    pylab.title('Time = '+str(time)+'hr on Oct 18, 2002')

    plt.subplot(1,2,2)
    plt.quiver(0.5*uvel_masked[time+1,level,imin:imax-1,jmin:jmax-1]+
           0.5*uvel_masked[time+1,level,imin+1:imax+1-1,jmin:jmax-1],
           0.5*vvel_masked[time+1,level,imin:imax-1,jmin:jmax-1]+
           0.5*vvel_masked[time+1,level,imin:imax-1,jmin+1:jmax+1-1],color='b',scale=4,pivot='middle')
    pylab.title('Time = '+str(time+1)+'hr on Oct 18, 2002')
    tempPlot = temp[time+1,level,:,:]
    land=tempPlot
    mu =land != 0
    land= np.ma.array(land,mask=mu)
    pylab.pcolor(np.arange(1,jmax-jmin)-1,np.arange(1,imax-imin)-1,
             land[imin+1:imax,jmin+1:jmax],vmin=lmin,vmax=lmax,cmap='hot')
```
