Notebook to show the deep incoming current on the east side of the southern Strait of Georgia.
These are <b>Preliminary</b> results.

```python
# import our goodies
%matplotlib inline
from matplotlib import pylab
import matplotlib.pyplot as plt
import netCDF4 as NC
import numpy as np
from salishsea_tools import viz_tools
```

```python
# Choose our results dates
# These results are forced with full tides, atmospheric forcing on a 33 km grid
# Rivers and ssh at Juan de Fuca are monthly climatologies
# TS at Juan de Fuca is weekly climatology
# boundary conditions are just wrong at the North
date1 = 'Jan 20, 2003'
time1 = 9
name1 = '11jan20jan/SalishSea_1d_20030111_20030120_grid_'

date2 = 'Jan 30, 2003'
time2 = 9
name2 = '21jan30jan/SalishSea_1d_20030121_20030130_grid_'
resultsloc = '/ocean/dlatorne/MEOPAR/SalishSea/results/spin-up/'
```

```python
# read in the data and mask it
fU1 = NC.Dataset(resultsloc+name1+'U.nc','r')
uvel1 = fU1.variables['vozocrtx'][time1,:]
mu = uvel1 == 0
uvel1_masked = np.ma.array(uvel1,mask=mu)
fV1 = NC.Dataset(resultsloc+name1+'V.nc','r')
vvel1 = fV1.variables['vomecrty'][time1,:]
depth = fV1.variables['depthv'][:]
mv = vvel1 == 0
vvel1_masked = np.ma.array(vvel1,mask=mv)
print uvel1.shape, time1
```

```python
# read in the second time stamp data, and mask it
fU2 = NC.Dataset(resultsloc+name2+'U.nc','r')
uvel2 = fU2.variables['vozocrtx'][time2,:]
mu = uvel2 == 0
uvel2_masked = np.ma.array(uvel2,mask=mu)
fV2 = NC.Dataset(resultsloc+name2+'V.nc','r')
vvel2 = fV2.variables['vomecrty'][time2,:]
mv = vvel2 == 0
vvel2_masked = np.ma.array(vvel2,mask=mv)
print uvel2.shape, time2
```

```python
# Show domain and location of the cross-section
grid = NC.Dataset('../../nemo-forcing/grid/bathy_meter_SalishSea2.nc')
bathy = grid.variables['Bathymetry'][:]
fig, ax = plt.subplots(1, 1, figsize=(10, 8))
viz_tools.set_aspect(ax)
cmap = plt.get_cmap('winter_r')
cmap.set_bad('burlywood')
mesh = ax.pcolormesh(bathy, cmap=cmap)
fig.colorbar(mesh)
plt.plot((320,240),(424,424),'-r')
plt.axis((0, bathy.shape[1], 0, bathy.shape[0]))
plt.title('Domain, Red Line Shows Cross-section')
```

```python
# plot velocity cross-section
jval = 424; imin=240; imax=320; id=37
vmax1 = np.max(vvel1_masked[:,jval,imin:imax])
vmax2 = np.max(vvel2_masked[:,jval,imin:imax])
vmin1 = np.min(vvel1_masked[:,jval,imin:imax])
vmin2 = np.min(vvel2_masked[:,jval,imin:imax])
vmax = max(vmax1,vmax2,-vmin1,-vmin2)
cmap = plt.get_cmap('bwr')
cmap.set_bad('burlywood')

plt.figure(figsize=(15,10))
plt.subplot(1,2,1)
plt.pcolormesh(np.arange(imin,imax),-depth[:id],vvel1_masked[:id,jval,imin:imax],
               cmap=cmap,vmax=vmax,vmin=-vmax)
plt.title(date1)
plt.plot(282,-170,'*',267,-300,'*')
plt.colorbar()

plt.subplot(1,2,2)
plt.pcolormesh(np.arange(imin,imax),-depth[:id],vvel2_masked[:id,jval,imin:imax],
               cmap=cmap,vmax=vmax,vmin=-vmax)
plt.title(date2)
plt.plot(282,-170,'*',267,-300,'*')
plt.colorbar()
print "Stars mark the approximate location of the two VENUS nodes"
print "One day average, 'North' Velocity (m/s)"
```

```python
# Show domain and location of the velocity map
fig, ax = plt.subplots(1, 1, figsize=(10, 8))
viz_tools.set_aspect(ax)
cmap = plt.get_cmap('winter_r')
cmap.set_bad('burlywood')
mesh = ax.pcolormesh(bathy, cmap=cmap)
fig.colorbar(mesh)
jmin = 240
jmax = 320
imin = 420-40
imax = 420+40
plt.plot((jmin,jmax),(imin,imin),'-b')
plt.plot((jmin,jmax),(imax,imax),'-b')
plt.plot((jmin,jmin),(imin,imax),'-b')
plt.plot((jmax,jmax),(imin,imax),'-b')
plt.axis((0, bathy.shape[1], 0, bathy.shape[0]))
plt.title('Domain, Blue Box Shows Plot Region')
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
viz_tools.set_aspect(ax)
level = 25
plt.pcolormesh(1-mv[level,imin:imax,jmin:jmax],cmap='hot')
plt.quiver(0.5*uvel2_masked[level,imin:imax,jmin:jmax]+0.5*uvel2_masked[level,imin-1:imax-1,jmin:jmax],
           0.5*vvel2_masked[level,imin:imax,jmin:jmax]+0.5*vvel2_masked[level,imin:imax,jmin-1:jmax-1],color='r')
plt.quiver(0.5*uvel1_masked[level,imin:imax,jmin:jmax]+0.5*uvel1_masked[level,imin-1:imax-1,jmin:jmax],
           0.5*vvel1_masked[level,imin:imax,jmin:jmax]+0.5*vvel1_masked[level,imin:imax,jmin-1:jmax-1],color='b')
plt.title('One day average, velocity, in red day,'+date2+' in blue day,'+date1)
print 'depth =',depth[level],' m'
```
