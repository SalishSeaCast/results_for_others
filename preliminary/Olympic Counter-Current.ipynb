Very preliminary results on currents in Juan de Fuca

# Juan de Fuca Strait Currents in August 2002

```python
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import pylab
import netCDF4 as NC
import numpy as np
from salishsea_tools import viz_tools
```

```python
# get the grid
grid = NC.Dataset('../../nemo-forcing/grid/bathy_meter_SalishSea2.nc')
lats = grid.variables['nav_lat']
lons = grid.variables['nav_lon']
```

```python
# Set up the time slice here.  date is used for headings,  time is
# the time counter in the file, name is the files, with resultsloc its directory

date = 'Aug 3'
time = 4
name = '30jul8aug/SalishSea_1d_20030730_20030808_grid_'
resultsloc = '/ocean/dlatorne/MEOPAR/SalishSea/results/spin-up/'
```

```python
# Tracer file : salinity, depth and mask it
fT1 = NC.Dataset(resultsloc+name+'T.nc','r')
sal = fT1.variables['vosaline'][time,:]
depth = fT1.variables['deptht'][:]
m = sal == 0
sal_masked = np.ma.array(sal,mask=m)
```

```python
# Velocity files
fU1 = NC.Dataset(resultsloc+name+'U.nc','r')
uvel = fU1.variables['vozocrtx'][time,:]
mu = uvel == 0
uvel_masked = np.ma.array(uvel,mask=mu)
fV1 = NC.Dataset(resultsloc+name+'V.nc','r')
vvel = fV1.variables['vomecrty'][time,:]
mv = vvel == 0
vvel_masked = np.ma.array(vvel,mask=mv)
```

```python
level = 15
im = 300; ix=476; jx = 100
fig, ax = plt.subplots(1,1,figsize=(12,5))
cmap = plt.get_cmap('spectral')
cmap.set_bad('burlywood')
array = uvel_masked[level,im:ix,:jx]-vvel_masked[level,im:ix,:jx]
# first one
viz_tools.set_aspect(ax, coords='map', lats=lats)
mesh = ax.pcolormesh(lons[im:ix,:jx],lats[im:ix,:jx],array,cmap=cmap)
viz_tools.plot_coastline(ax, grid, coords='map')
ax.set_title(date)
cbar=fig.colorbar(mesh)
#ax.contour(array,0, colors=('black'))
cbar.add_lines(CS)
plt.axis((-125, -123., 48, 49))
ipos = 20
j1 = 355; j2 = 435
plt.plot((lons[j1,ipos],lons[j2+5,ipos]),(lats[j1,ipos],lats[j2+5,ipos]),'-k')

# Add axis labels
ax.set_xlabel('{longitude.long_name} [{longitude.units}]'.format(longitude=lons))
ax.set_ylabel('{latitude.long_name} [{latitude.units}]'.format(latitude=lats))
cbar.set_label('Inflow [m/s]')
print "At depth = ", depth[level], 'm'
```

```python
# vertical cross-section
id = 34
plt.figure(figsize=(8,5))
plt.subplot(1,2,1)
plt.pcolormesh(np.arange(0,80),-depth[:id],uvel_masked[:id,j1:j2,ipos]-
               vvel_masked[:id,j1:j2,ipos],vmin=-0.5, vmax=0.5, cmap = 'bwr')
plt.title(date+' Inflow Velocity')
plt.colorbar()
plt.ylim(-300,0)
CS=pylab.contour(np.arange(0,80),-depth[:id],uvel_masked[:id,j1:j2,ipos]-
                 vvel_masked[:id,j1:j2,ipos],(0), colors='black')
pylab.clabel(CS, fontsize=9, inline=1)
plt.subplot(1,2,2)
plt.pcolormesh(np.arange(0,80),-depth[:id],sal_masked[:id,j1:j2,ipos-1],vmax=33.6,vmin=29.5)
plt.title(date+' Salinity')
plt.colorbar()
print "Cross-sections through line marked on map above"
```

Maximum inward velocity is 0.28 m/s.

Note: velocities are not de-tided, they are simply daily averages.
